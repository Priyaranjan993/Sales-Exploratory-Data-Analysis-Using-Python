# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
for i in ax.containers:
    ax.bar_label(i,fmt='%.0f')
```

```python
sales_gen=df.groupby('Gender', as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
ae=sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python


# Assuming you have already calculated the sales_gen dataframe



ae = sns.barplot(x='Gender', y='Amount', data=sales_gen)
for container in ae.containers:
    ae.bar_label(container, fmt='%.0f')  # Format the labels as floating point numbers with 2 decimal places
    
plt.show()

```

from the above graph we can see that the most of buyers are female even purchasing power of female greater than male

# Age

```python
age_grp=sns.countplot(x='Age Group',data=df,hue='Gender')
age_grp.set(title='Age group and gender orders')
for i in age_grp.containers:
     age_grp.bar_label(i)
```

# from the above grap we can see  most of buyer grops are 26-35 females

# state

```python
sales_state= df.groupby(["State"],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x='State',y='Orders',data=sales_state)
plt.title('State wise Total Orders',fontsize=35)
```

```python
sales_Amt= df.groupby(["State"],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)

sns.set(rc={'figure.figsize':(25,7)})
b=sns.barplot(x='State',y='Amount',data=sales_Amt)
plt.title('State wise Total Amount',fontsize=35)
for container in b.containers:
    b.bar_label(container, fmt='%.0f') 
```

# from the above graph most of the oders & total sales from UttarPradesh,Maharashtra & Karnatka


```python
age_grp=sns.countplot(x='Marital_Status',data=df,hue='Gender')
age_grp.set(title='Marital_Status and gender orders')
sns.set(rc={'figure.figsize':(8,3)})
for i in age_grp.containers:
     age_grp.bar_label(i)
```

# in the above graph female marrried persons are purchase more

```python
age_grp=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(25,5)})
plt.title('Occupation and gender orders',fontsize=35)
for i in age_grp.containers:
     age_grp.bar_label(i)
```

```python
pucharsing_amount_occupations= df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
c=sns.barplot(x='Occupation',y='Amount',data=pucharsing_amount_occupations)
plt.title('occupaton wise total Amount spet',fontsize=30)
for i in c.containers:
    c.bar_label(i,fmt='%.0f')
```

```python
Category_product=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(25,5)})
plt.title('Product_category',fontsize=35)
for i in Category_product.containers:
     Category_product.bar_label(i)
```

Total Number of Orders in prodct Categorywise. we found clothing and apparel and food are most buying category


```python
Categorywise_amount_spent= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
z=sns.barplot(x='Product_Category',y='Amount',data=Categorywise_amount_spent)
plt.title('Categorywise_amount_spent',fontsize=30)
for i in z.containers:
    z.bar_label(i,fmt='%.0f')
```

# from the above graph food orders more and spent more amount

```python
Sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
z=sns.barplot(x='Product_ID',y='Orders',data=Sales_state)
plt.title('Sales_state',fontsize=30)
for i in z.containers:
    z.bar_label(i,fmt='%.0f')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
